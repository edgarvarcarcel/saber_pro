# **4. Entrenamiento de modelos**
En esta sección se entrenaran los siguientes modelos:

1. Extreme Gradient Boosting (XGBoost).
2. Hist Gradient Boosting,
3. Perceptrón Multicapa de la librería de Scikit Learn.
4. Fully Conected Neural Network de la librería Pytorch con capas de normalización y de dropout.
5. Random Forest.
6. Máquinas de Vectores de Soporte

Los modelos se entrenarán usando el split de entrenamiento generado anteriormente, haciendo uso de validación cruzada con 5-folds y se evaluarán sobre el subset de validación del fold, y el split de prueba para obtener una distribución de métricas al final del proceso.

```python
# Definir arquitectura del modelo de Pytorch
class FullyConnectedModel(nn.Module):
    def __init__(self, input_size, hidden_sizes, output_size, dropout_prob=0.5):
        super(FullyConnectedModel, self).__init__()
        self.input_size = input_size
        self.hidden_sizes = hidden_sizes
        self.output_size = output_size
        self.dropout_prob = dropout_prob

        layers = []
        # Crear capas ocultas
        prev_size = input_size
        for hidden_size in hidden_sizes:
            layers.append(nn.Linear(prev_size, hidden_size))
            layers.append(nn.BatchNorm1d(hidden_size))
            layers.append(nn.Dropout(p=self.dropout_prob))
            layers.append(nn.ReLU())
            prev_size = hidden_size

        # Capa de salida con activación linear
        layers.append(nn.Linear(prev_size, output_size))
        layers.append(nn.Linear(output_size , 1))

        self.layers = nn.Sequential(*layers)

    def forward(self, x):
        return self.layers(x)

# Define function to save pytorch model for early stopping
def checkpoint(model, filename):
    torch.save(model.state_dict(), filename)

# Define function to load best early stopping pytorch model to continue with the evaluation
def resume(model, filename):
    model.load_state_dict(torch.load(filename))
```

```python
# XGBoost
results_xgb_10000 = {}
results_test_xgb_10000 = {}
results_train_xgb_10000 = {}
model_name = 'XGBoost'
model_parameters = {'n_estimators' : [100 , 200 , 300],
                    'max_features' : ['auto', 'sqrt', 'log2'],
                    'booster' : ['gbtree', 'gblinear']}
# Parameters for saving results
start_2 = time.time()
aux_save = {}
fold_f1_test = []
fold_auc_test = []
fold_acc_test = []
fold_precision_test = []
fold_recall_test = []
fold_specificity_test = []
for i, (train_index, test_index) in enumerate(skf.split(X_gaussian, Y_gaussian)):
  print('#' * 50)
  print('Iteration' , i + 1 , 'of' , 5)
  print('#' * 50)

  # Re define train and val
  x_train = X_gaussian.reset_index(drop = True).loc[train_index]
  y_train = Y_gaussian.reset_index(drop = True).loc[train_index]

  x_val = X_gaussian.reset_index(drop = True).loc[test_index]
  y_val = Y_gaussian.reset_index(drop = True).loc[test_index]

  # Create and train model
  print('Training Model')
  base_model = XGBRegressor()
  model = GridSearchCV(base_model,
                       model_parameters,
                       verbose = 10,
                       cv = 2,
                       scoring = 'r2')
  model.fit(x_train , y_train)

  # Predictions
  print('Making predictions on train set')
  predictions = model.predict(x_train)

  # Compute metrics for test set
  rmse = mean_squared_error(y_train.values , predictions) ** 0.5
  mae = mean_absolute_error(y_train.values , predictions)
  mape = mean_absolute_percentage_error(y_train.values , predictions)
  print('RMSE:' , rmse)
  print('MAE:' , mae)
  print('MAPE:' , mape)

  # Save results
  results_train_xgb_10000[i+1] = {'RMSE' : [rmse],
                                  'MAE' : [mae],
                                  'MAPE' : [mape]}

  # Predictions
  print('Making predictions on val set')
  predictions = model.predict(x_val)

  # Compute metrics for test set
  print('Computing Metrics for test set')
  rmse = mean_squared_error(y_val.values , predictions) ** 0.5
  mae = mean_absolute_error(y_val.values , predictions)
  mape = mean_absolute_percentage_error(y_val.values , predictions)
  print('RMSE:' , rmse)
  print('MAE:' , mae)
  print('MAPE:' , mape)

  # Save results
  results_xgb_10000[i+1] = {'RMSE' : [rmse],
                            'MAE' : [mae],
                            'MAPE' : [mape]}


  # Predictions
  print('Making predictions on test set')
  predictions = model.predict(X_test)

  # Compute metrics for test set
  print('Computing Metrics for test set')
  rmse = mean_squared_error(Y_test.values , predictions) ** 0.5
  mae = mean_absolute_error(Y_test.values , predictions)
  mape = mean_absolute_percentage_error(Y_test.values , predictions)
  print('RMSE:' , rmse)
  print('MAE:' , mae)
  print('MAPE:' , mape)

  results_test_xgb_10000[i + 1] = {'RMSE' : [rmse],
                                  'MAE' : [mae],
                                  'MAPE' : [mape],
                                  'Parameters' : [model.best_params_]}
```

## **4.3 Pytorch Neural Network**

```python
# Neural Network
BATCH_SIZE = 16
results_pytorch = {}
results_test_pytorch = {}
results_train_pytorch = {}
model_name = 'Neural Network'
model_parameters = [[100 , 100 , 100],
                    [100 , 50 , 100],
                    [50 , 100 , 50],
                    [100 , 50 , 100],
                    [128 , 1024 , 512 , 128]]
for i, (train_index, test_index) in enumerate(skf.split(X_gaussian, Y_gaussian)):
  print('#' * 50)
  print('Iteration' , i + 1 , 'of' , 5)
  print('#' * 50)

  # Re define train and val
  x_train = X_gaussian.reset_index(drop = True).loc[train_index]
  y_train = Y_gaussian.reset_index(drop = True).loc[train_index]

  x_val = X_gaussian.reset_index(drop = True).loc[test_index]
  y_val = Y_gaussian.reset_index(drop = True).loc[test_index]

  # Convert data to Pytorch tensors
  x_train_tensor = torch.FloatTensor(x_train.values)
  x_test_tensor = torch.FloatTensor(X_test.values)
  x_val_tensor = torch.FloatTensor(x_val.values)

  y_train_tensor = torch.FloatTensor(y_train.values)
  y_test_tensor = torch.FloatTensor(Y_test.values)
  y_val_tensor = torch.FloatTensor(y_val.values)

  dataset = torch.utils.data.TensorDataset(x_train_tensor, y_train_tensor)
  dataloader = torch.utils.data.DataLoader(dataset, batch_size = BATCH_SIZE, shuffle=True)
  dataset_size = len(dataloader.dataset)

  # Define the model
  for j in range(len(model_parameters)):
    print('Training with hidden sizes -->' , model_parameters[j])
    input_size = x_train.shape[1]
    hidden_sizes = model_parameters[j]
    output_size = 3
    dropout_prob = 0.2
    model = FullyConnectedModel(input_size, hidden_sizes, output_size, dropout_prob)
    # Define the loss function and optimizer
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    # Define the early stopping criteria
    early_stop_thresh = 50
    best_val_loss = 1e10
    best_epoch = -1

    # Define Scheduler for Learning Rate
    scheduler = torch.optim.lr_scheduler.LinearLR(optimizer, start_factor=1.0, end_factor=0.5, total_iters=30)

    # Training loop
    print('Training Model')
    num_epochs = 500
    aux_val_loss = []
    train_loss_history = []
    val_loss_history = []
    test_loss_history = []
    for epoch in range(num_epochs):
      print('-' * 50)
      print(f"Epoch {epoch + 1}")
      for id_batch, (x_batch, y_batch) in enumerate(dataloader):
        if x_batch.shape[0] > 1:
          optimizer.zero_grad()
          outputs = model(x_batch)
          loss = criterion(outputs, y_batch)
          loss.backward()
          optimizer.step()
      # Scheduler learning rate
      before_lr = optimizer.param_groups[0]["lr"]
      scheduler.step()
      after_lr = optimizer.param_groups[0]["lr"]
      # Eval model for early stop
      model.eval()
      test_outputs = model(x_test_tensor)
      val_loss = criterion(test_outputs , y_test_tensor)
      val_outputs = model(x_val_tensor)
      test_loss = criterion(val_outputs , y_val_tensor)
      aux_val_loss.append(val_loss.item())
      test_loss_history.append(val_loss.item())
      train_loss_history.append(loss.item())
      val_loss_history.append(test_loss.item())
      if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_epoch = epoch
        checkpoint(model, "best_model.pth")
      elif epoch - best_epoch > early_stop_thresh:
          print("Early stopped training at epoch %d" % epoch)
          break  # terminate the training loop
      print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {loss.item():.4f} , Val Loss: {val_loss.item():.4f}')
      print('-' * 50)

    # Evaluation on train set
    # Load best model
    print('Making predictions')
    print('#' * 25)
    print('Evaluation on train set')
    resume(model, "best_model.pth")
    model.eval()
    with torch.no_grad():
        test_outputs = model(x_train_tensor)
        test_outputs = test_outputs.squeeze()
    # Compute metrics for test set
    print('Computing Metrics for train set')
    predictions = test_outputs.numpy()
    rmse = mean_squared_error(y_train.values , predictions) ** 0.5
    mae = mean_absolute_error(y_train.values , predictions)
    mape = mean_absolute_percentage_error(y_train.values , predictions)
    print('RMSE:' , rmse)
    print('MAE:' , mae)
    print('MAPE:' , mape)

    # Save results
    results_train_pytorch[str(i+1) + "_" + str(j)] = { 'RMSE' : [rmse],
                                                      'MAE' : [mae],
                                                       'MAPE' : [mape],
                                          'Parameters' : [model_parameters[j]]}

    # Evaluation on val set
    # Load best model
    print('Making predictions')
    print('#' * 25)
    print('Evaluation on val set')
    resume(model, "best_model.pth")
    model.eval()
    with torch.no_grad():
        test_outputs = model(x_val_tensor)
        test_outputs = test_outputs.squeeze()
    # Compute metrics for test set
    print('Computing Metrics for val set')
    predictions = test_outputs.numpy()
    rmse = mean_squared_error(y_val.values , predictions) ** 0.5
    mae = mean_absolute_error(y_val.values , predictions)
    mape = mean_absolute_percentage_error(y_val.values , predictions)
    print('RMSE:' , rmse)
    print('MAE:' , mae)
    print('MAPE:' , mape)

    # Save results
    results_pytorch[str(i+1) + "_" + str(j)] = {'RMSE' : [rmse],
                                                'MAE' : [mae],
                                                'MAPE' : [mape],
                                          'Parameters' : [model_parameters[j]]}

    # Evaluation on test set
    # Load best model
    print('Making predictions')
    print('#' * 25)
    print('Evaluation on test set')
    resume(model, "best_model.pth")
    model.eval()
    with torch.no_grad():
        test_outputs = model(x_test_tensor)
        test_outputs = test_outputs.squeeze()
    # Compute metrics for test set
    print('Computing Metrics for test set')
    predictions = test_outputs.numpy()
    rmse = mean_squared_error(Y_test.values , predictions) ** 0.5
    mae = mean_absolute_error(Y_test.values , predictions)
    mape = mean_absolute_percentage_error(Y_test.values , predictions)
    print('RMSE:' , rmse)
    print('MAE:' , mae)
    print('MAPE:' , mape)

    # Save results
    results_test_pytorch[str(i+1) + "_" + str(j)] = {'RMSE' : [rmse],
                                                     'MAE' : [mae],
                                                     'MAPE' : [mape],
                                          'Parameters' : [model_parameters[j]]}
    # plot training and validation loss
    plt.plot(train_loss_history, label="Train Loss")
    plt.plot(val_loss_history, label="Val Loss")
    plt.plot(test_loss_history, label="Test Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()
```

```python
# XGBoost
results_xgb_10000 = {}
results_test_xgb_10000 = {}
results_train_xgb_10000 = {}
model_name = 'SVM'
model_parameters = {'kernel' : ['linear' , 'poly'],
                    'C' : [0.5 , 1 ,2]}
# Parameters for saving results
start_2 = time.time()
aux_save = {}
fold_f1_test = []
fold_auc_test = []
fold_acc_test = []
fold_precision_test = []
fold_recall_test = []
fold_specificity_test = []
for i, (train_index, test_index) in enumerate(skf.split(X_gaussian, Y_gaussian)):
  print('#' * 50)
  print('Iteration' , i + 1 , 'of' , 5)
  print('#' * 50)

  # Re define train and val
  x_train = X_gaussian.reset_index(drop = True).loc[train_index]
  y_train = Y_gaussian.reset_index(drop = True).loc[train_index]

  x_val = X_gaussian.reset_index(drop = True).loc[test_index]
  y_val = Y_gaussian.reset_index(drop = True).loc[test_index]

  # Create and train model
  print('Training Model')
  base_model = SVR()
  model = GridSearchCV(base_model,
                       model_parameters,
                       verbose = 10,
                       cv = 2,
                       scoring = 'r2')
  model.fit(x_train.values , y_train)

  # Predictions
  print('Making predictions on train set')
  predictions = model.predict(x_train.values)

  # Compute metrics for test set
  rmse = mean_squared_error(y_train.values , predictions) ** 0.5
  mae = mean_absolute_error(y_train.values , predictions)
  mape = mean_absolute_percentage_error(y_train.values , predictions)
  print('RMSE:' , rmse)
  print('MAE:' , mae)
  print('MAPE:' , mape)

  # Save results
  results_train_xgb_10000[i+1] = {'RMSE' : [rmse],
                                  'MAE' : [mae],
                                  'MAPE' : [mape]}

  # Predictions
  print('Making predictions on val set')
  predictions = model.predict(x_val.values)

  # Compute metrics for test set
  print('Computing Metrics for test set')
  rmse = mean_squared_error(y_val.values , predictions) ** 0.5
  mae = mean_absolute_error(y_val.values , predictions)
  mape = mean_absolute_percentage_error(y_val.values , predictions)
  print('RMSE:' , rmse)
  print('MAE:' , mae)
  print('MAPE:' , mape)

  # Save results
  results_xgb_10000[i+1] = {'RMSE' : [rmse],
                            'MAE' : [mae],
                            'MAPE' : [mape]}


  # Predictions
  print('Making predictions on test set')
  predictions = model.predict(X_test.values)

  # Compute metrics for test set
  print('Computing Metrics for test set')
  rmse = mean_squared_error(Y_test.values , predictions) ** 0.5
  mae = mean_absolute_error(Y_test.values , predictions)
  mape = mean_absolute_percentage_error(Y_test.values , predictions)
  print('RMSE:' , rmse)
  print('MAE:' , mae)
  print('MAPE:' , mape)

  results_test_xgb_10000[i + 1] = {'RMSE' : [rmse],
                                  'MAE' : [mae],
                                  'MAPE' : [mape],
                                  'Parameters' : [model.best_params_]}
```
