```python
# Check for outliers on numeric columns
outliers = {}
def detect_outliers_zscore(df, threshold=3):
  z_scores = np.abs(stats.zscore(df))
  return df.reset_index(drop = True).loc[np.where(z_scores > threshold)]
for i in num_columns:
  print('Checking for outliers on -->' , i)
  outliers[i] = detect_outliers_zscore(df[i])
# Show outliers
for i in outliers.keys():
  print('-' * 50)
  print('Outliers for feature -->' , i)
  print(outliers[i])
```

## **1.1 Scaling + One Hot Encoding + Ordinal Encoding**

```python
# Create scaler and one hot encoding obects
numeric_scaler = MinMaxScaler()
cat_encoder = OneHotEncoder()
# Fit objects
numeric_scaler.fit(df[num_columns])
cat_encoder.fit(df[cat_columns])
# Transform data
aux_numeric = pd.DataFrame(numeric_scaler.transform(df[num_columns]) , columns = df[num_columns].columns.tolist())
aux_categorical = pd.DataFrame.sparse.from_spmatrix(cat_encoder.transform(df[cat_columns]) , columns = cat_encoder.get_feature_names_out().tolist())
# Concat data
df_scaled = pd.concat([aux_numeric,
                       aux_categorical,
                       df[ordinal_columns],
                       df[TARGET]] , axis = 1)
df_scaled
```

## **1.2 Add Gaussian Noise**

```python
# Create gaussian noise for numerical features
std_dev = 0.1  # You can change this parameter
noise = np.random.normal(loc=0,
                         scale=std_dev,
                         size = (df_scaled.shape[0] , df_scaled.drop(columns = TARGET).shape[1]))
# Create New DataFrame for adding gaussian noise
non_numeric_columns = [i for i in df_scaled.columns.tolist() if i not in num_columns and i not in TARGET and i not in ordinal_columns]
aux = pd.concat([df_scaled.drop(columns = TARGET) + noise,
                 df_scaled[TARGET]] , axis = 1)
df_gaussian_noise = pd.concat([df_scaled,
                               aux] , axis = 0)

# Load the original dataframe for indexing the group
df_original = pd.concat([train_set,
                          test_set] , axis = 0).reset_index(drop = True)
# Put the group column to create the index for the augmented data
df_aux = pd.concat([df_original,
                    df_original] , axis = 0)
df_scaled_noise = pd.concat([df_gaussian_noise.reset_index(drop = True),
                             df_aux['group'].reset_index(drop = True)] , axis = 1).reset_index(drop = True)
# Extract the index from de gaussian noise dataframe to know which samples are from each group
index_augmented_test = df_scaled_noise[df_scaled_noise['group'].isin( group_for_test)].index

# Mask nan values
df_scaled_noise = df_scaled_noise.fillna(-1)
# Drop group column since we are not going to use it
df_scaled_noise = df_scaled_noise.drop(columns = ['group'])
# Show data frame that has augmented data (noise on categorical and numerical), scaled and one hot encoding
df_scaled_noise
```

# **2. Model Training and Evaluation**

```python
# Define functions for evaluation metrics
def calculate_confusion_matrix(true_labels, predicted_labels):
    # Calculate confusion matrix
    tn, fp, fn, tp = confusion_matrix(true_labels, predicted_labels).ravel()

    return tn, tp, fp, fn

# Define function that takes confusion matrix an compute required metrics
def get_metrics(TN,TP,FP,FN):

    acc = (TN+TP)/(TN+TP+FN+FP)
    precision = TP / (TP + FP)

    # Calculate recall
    recall = TP / (TP + FN)
    # Calulcate specificity
    specificity = TN / (TN + FP)

    # Calculate False Negative Rate
    FNR = FN / (FN + TP)

    # Calculate F1 score
    f1 = 2 * (precision * recall) / (precision + recall)

    return acc, precision, recall, f1 , specificity , FNR
```

```python
# Function to smooth the ROC curve using a moving average
def smooth_roc_curve(fpr, tpr, window_size=5):
    df = pd.DataFrame({'fpr': fpr, 'tpr': tpr})
    df_smoothed = df.rolling(window=window_size).mean().dropna()
    # Add (1,1) point at the end
    df_smoothed = pd.concat([df_smoothed,
                             pd.DataFrame({'fpr' : [1] , 'tpr' : [1]})] , axis = 0)
    return df_smoothed['fpr'].values, df_smoothed['tpr'].values


# Function to smooth the Precision Recall curve using a moving average
def smooth_precision_curve(precision, recall, window_size=5):
    df = pd.DataFrame({'precision': precision, 'recall': recall})
    df_smoothed = df.rolling(window=window_size).mean().dropna()
    # Add (1,0) point at the end
    df_smoothed = pd.concat([df_smoothed,
                             pd.DataFrame({'precision' : [1] , 'recall' : [0]})] , axis = 0)
    return df_smoothed['precision'].values, df_smoothed['recall'].values
```

```python
# Define Seed for train_test split
SEED = 1234
# Define number of folds
NUMBER_OF_FOLDS = 5
```

```python
# Create test set of group 4
df_test_group_4 = df_scaled_noise.loc[index_augmented_test]
# Define X and Y of group 4
X_test_group_4 = df_test_group_4.drop(columns = TARGET)
Y_test_group_4 = df_test_group_4[TARGET]
# Show the data frame that we are going to use for test set of group 4
df_test_group_4
```

```python
# Select group 1, 2 and 3 and split in a random train / test
df_1_2_3 = df_scaled_noise[~df_scaled_noise.index.isin(index_augmented_test.tolist())]
X = df_1_2_3.drop(columns = TARGET)
Y = df_1_2_3[TARGET]
# Perform random test split
X_train , X_test , Y_train , Y_test = train_test_split(X,
                                                       Y,
                                                       test_size = 0.1,
                                                       random_state = SEED,
                                                       stratify = Y)
# Stratified K fold cross validation
skf = RepeatedStratifiedKFold(n_splits = NUMBER_OF_FOLDS
                              ,
                              n_repeats = 1 ,
                              random_state = SEED)
```

```python
# Define parameters for saving the information
results_test_random = pd.DataFrame()
results_train = pd.DataFrame()
results_test_holdout = pd.DataFrame()
results_val = pd.DataFrame()
model_name = 'GLM'
model_parameters = {'C' : [0.25 , 0.5 , 0.75 , 1 , 2],
                    'class_weight' : ['balanced' , None],
                    'penalty' : ['l1' , 'l2' , 'elasticnet' , None]}
# Extract total combinations
num_combinations = len(list(product(*model_parameters.values())))
all_combinations = [
    {key: value for key, value in zip(model_parameters.keys(), combo)}
    for combo in product(*model_parameters.values())]
# Loop throught folds
for i, (train_index, test_index) in enumerate(skf.split(X_train, Y_train)):
  print('#' * 50)
  print('Iteration' , i + 1 , 'of' , NUMBER_OF_FOLDS)
  print('#' * 50)
  # Re define train and val
  x_train = X_train.reset_index(drop = True).loc[train_index]
  y_train = Y_train.reset_index(drop = True).loc[train_index]

  x_val = X_train.reset_index(drop = True).loc[test_index]
  y_val = Y_train.reset_index(drop = True).loc[test_index]
  for ii in range(len(all_combinations)):
    print('-' * 25)
    print('Combination' , ii + 1, 'of' , num_combinations)
    print('Parameters -->' , all_combinations[ii])
    print('-' * 25)
    # Create and train model
    model = LogisticRegression(solver = 'saga' , **all_combinations[ii])
    try:
      model.fit(x_train , y_train)
      # Predictions on train set
      print('Making predictions on train set')
      predictions = model.predict(x_train)
      predictions_proba = model.predict_proba(x_train)[: , 1]
      # Compute metrics for train set
      print('Computing Metrics for train set')
      try:
          roc_auc = roc_auc_score(y_train.values , predictions_proba)
      except:
        roc_auc = 0
      TN, TP, FP, FN = calculate_confusion_matrix(y_train.values,
                                                  predictions)
      acc, precision, recall, f1 , specificity, fnr = get_metrics(TN,
                                                                  TP,
                                                                  FP,
                                                                  FN)
      f1 = f1_score(y_train.values , predictions , average = 'micro')
      f2 = f1_score(y_train.values , predictions , average = 'macro')
      print(confusion_matrix(y_train.values , predictions))
      print('Accuracy:' , acc)
      print('Precision:' , precision)
      print('Recall:' , recall)
      print('False Negative Ratio:' , fnr)
      print('F1 Micro:' , f1)
      print('F1 Macro:' , f2 )

      # Save results of train set
      aux_train =  pd.DataFrame({'Accuracy' : [acc],
                        'Precision' : [precision],
                        'Recall' : [recall],
                        'F1_Micro' : [f1],
                        'F1_Macro' : [f2],
                        'Specificity' : [specificity],
                        'False Negative Ratio' : [fnr],
                        'Fold' : [i],
                        'model_parameters' :  [all_combinations[ii]]})
      results_train = pd.concat([results_train,
                                aux_train])

      print('Making predictions on val set from cross validation')
      predictions = model.predict(x_val)
      predictions_proba = model.predict_proba(x_val)[: , 1]
      # Compute metrics for train set
      print('Computing Metrics for val set from cross validation')
      try:
          roc_auc = roc_auc_score(y_val.values , predictions_proba)
      except:
        roc_auc = 0
      TN, TP, FP, FN = calculate_confusion_matrix(y_val.values,
                                                  predictions)
      acc, precision, recall, f1 , specificity, fnr = get_metrics(TN,
                                                                  TP,
                                                                  FP,
                                                                  FN)
      f1 = f1_score(y_val.values , predictions , average = 'micro')
      f2 = f1_score(y_val.values , predictions , average = 'macro')
      print(confusion_matrix(y_val.values , predictions))
      print('Accuracy:' , acc)
      print('Precision:' , precision)
      print('Recall:' , recall)
      print('False Negative Ratio:' , fnr)
      print('F1 Micro:' , f1)
      print('F1 Macro:' , f2 )

      # Save results of train set
      aux_val =  pd.DataFrame({'Accuracy' : [acc],
                        'Precision' : [precision],
                        'Recall' : [recall],
                        'F1_Micro' : [f1],
                        'F1_Macro' : [f2],
                        'Specificity' : [specificity],
                        'False Negative Ratio' : [fnr],
                        'Fold' : [i],
                        'model_parameters' :  [all_combinations[ii]]})
      results_val = pd.concat([results_val,
                                aux_val])
      # Predictions on RANDOM test set
      print('Making predictions on RANDOM test set')
      predictions = model.predict(X_test)
      predictions_proba = model.predict_proba(X_test)[: , 1]

      # Compute metrics for RANDOM test
      print('Computing Metrics for RANDOM test')
      try:
          roc_auc = roc_auc_score(Y_test.values , predictions_proba)
      except:
          roc_auc = 0
      TN, TP, FP, FN = calculate_confusion_matrix(Y_test.values,
                                                  predictions)
      acc, precision, recall, f1 , specificity, fnr = get_metrics(TN,
                                                                  TP,
                                                                  FP,
                                                                  FN)
      f1 = f1_score(Y_test.values , predictions , average = 'micro')
      f2 = f1_score(Y_test.values , predictions , average = 'macro')
      print('Results for' , model_name , 'in test set')
      print(confusion_matrix(Y_test.values , predictions))
      print('Accuracy:' , acc)
      print('Precision:' , precision)
      print('Recall:' , recall)
      print('False Negative Ratio:' , fnr)
      print('F1 Micro:' , f1)
      print('F1 Macro:' , f2 )

      # Save results from RANDOM test set
      aux_test_random =  pd.DataFrame({'Accuracy' : [acc],
                        'Precision' : [precision],
                        'Recall' : [recall],
                        'F1_Micro' : [f1],
                        'F1_Macro' : [f2],
                        'Specificity' : [specificity],
                        'False Negative Ratio' : [fnr],
                        'Fold' : [i],
                        'model_parameters' :  [all_combinations[ii]]})
      results_test_random = pd.concat([results_test_random,
                                      aux_test_random])

      # Predictions on holdout test set
      print('Making predictions on test set of group ' , group_for_test)
      predictions = model.predict(X_test_group_4)
      predictions_proba = model.predict_proba(X_test_group_4)[: , 1]

      # Compute metrics for test set
      print('Computing Metrics for test set of group ' , group_for_test)
      try:
          roc_auc = roc_auc_score(Y_test_group_4.values , predictions_proba)
      except:
          roc_auc = 0
      TN, TP, FP, FN = calculate_confusion_matrix(Y_test_group_4.values,
                                                  predictions)
      acc, precision, recall, f1 , specificity, fnr = get_metrics(TN,
                                                                  TP,
                                                                  FP,
                                                                  FN)
      f1 = f1_score(Y_test_group_4.values , predictions , average = 'micro')
      f2 = f1_score(Y_test_group_4.values , predictions , average = 'macro')
      print('Results for' , model_name , 'in test set')
      print(confusion_matrix(Y_test_group_4.values , predictions))
      print('Accuracy:' , acc)
      print('Precision:' , precision)
      print('Recall:' , recall)
      print('False Negative Ratio:' , fnr)
      print('F1 Micro:' , f1)
      print('F1 Macro:' , f2 )
      # Save results from holdout 4 TEST set
      aux_test_holdout =  pd.DataFrame({'Accuracy' : [acc],
                        'Precision' : [precision],
                        'Recall' : [recall],
                        'F1_Micro' : [f1],
                        'F1_Macro' : [f2],
                        'Specificity' : [specificity],
                        'False Negative Ratio' : [fnr],
                        'Fold' : [i],
                        'model_parameters' :  [all_combinations[ii]]})
      results_test_holdout = pd.concat([results_test_holdout,
                                      aux_test_holdout])
    except:
      print('Parameters non compatibles, ommiting iteration...')
```

```python
# Select best parameters
best_parameters = results_val.sort_values(by = 'F1_Macro' , ascending = False).reset_index(drop = True).loc[0 , :]['model_parameters']
print('Best parameters:' , best_parameters)
# Retrain model with best parameters
model = LogisticRegression(solver = 'saga' , **best_parameters)
model.fit(X_train , Y_train)
# Show model
model
```
